```python
from itertools import product

all = product(['C1','C2'], [i+1 for i in range(3)], ["Result"])
dims = ["Coins","Trials","Values"]
all = [{d:v for d,v in zip(dims, vals)} for vals in all]
input = "foobar"
for i in range(len(all)):
    print("Iter: ",i)
    qCoin.setValue(all[i], input[i])
    print(qCoin)
```
